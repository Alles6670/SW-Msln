```python
##시험 데이터 로드 

tst = pd.read_csv(tst_file, index_col = 0)
print(tst.shape)
tst.head()
```

```python
##학습/시험 데이터 결합

df = pd.concat([trn, tst], axis = 0)
    #trn과 tst를 결합(concat 함수)
    #axis 변수를 통해 결합 형태 정의 (ex. 0 = 위 아래로 결합을 하겠다, 1 = 옆으로 결합)
print(df.shape)
df.tail()
    #마지막 column class는 우리가 예측해야하는 값이기 때문에 Nan
```

```python
df.fillna(-1, inplace = True)
df.tail()
    #Nan 값을 대치하기 위한 fillna 함수
```

```python
##피쳐 변환


df['nObserve'].hist(bins = 30)
    #nObserve의 분포는 0에서 값이 많고 커질 수록 개수가 줄어듬(exponential distribution)
```

```python
df['nObserve'] = df['nObserve'].apply(np.log1p)
    #이런 경우엔 로그를 취하는게 보편적(apply(np.log)함수 : 칼럼 전체에 로그취함)
    #로그0은 무한대이므로 1 + 양수를 통해 -무한대 값이 안나오기 위해 np.log1p 
```

```python
df['nObserve'].hist(bins = 30)
    #최대값이 40 -> 35 정도로 낮춰짐
```

```python
#피쳐 생성 (또 다른 방법)


df['d_dered_u'] = df['dered_u'] - df['u']
df['d_dered_g'] = df['dered_g'] - df['g']
df['d_dered_r'] = df['dered_r'] - df['r']
df['d_dered_i'] = df['dered_i'] - df['i']
df['d_dered_z'] = df['dered_z'] - df['z']
df['d_dered_rg'] = df['dered_r'] - df['dered_g']
df['d_dered_ig'] = df['dered_i'] - df['dered_g']
df['d_dered_zg'] = df['dered_z'] - df['dered_g']
df['d_dered_ri'] = df['dered_r'] - df['dered_i']
df['d_dered_rz'] = df['dered_r'] - df['dered_z']
df['d_dered_iz'] = df['dered_i'] - df['dered_z']
df['d_obs_det'] = df['nObserve'] - df['nDetect']
    #칼럼은 사칙연산이 가능하기 때문에 각 칼럼의 연산값을 새로운 칼럼으로써 저장한다.
print(df.shape)
df.head()
```

```python
#피쳐 삭제
    #중복되는 쓸데없는 피쳐들을 삭제
    
    
df.corr().style.background_gradient()
    #상관계수를 확인하기위해 출력(df.corr()함수)
    #가장 높은 값은 짙은색, 가장 낮은 값은 연한색
    #어떤 값들이 상관계수들이 높은지 확인 가능 (ex. g와 dered_g가 상관계수가 높구나~)
    #g와 dered_g 둘중 하나만 사용해도 되겠구나 혹은 둘다 필요없다 ,airmass들은 거의 1에 근접하기때문에 1개만 사용가능
        #-> 이렇게 유추 가능, 이제 삭제 하자
```

```python
df.drop(['airmass_z', 'airmass_i', 'airmass_r', 'airmass_g', 'u', 'g', 'r', 'i', 'nDetect', 'd_dered_rg', 'd_dered_ri'], 
        axis=1, inplace=True)
    #drop함수를 통해 삭제, axis = 1 : 삭제할래, inplace = True : 데이터 frame 내에서 새로운 변수 생성없이 자체frame을 수정
print(df.shape)
df.head()
```

```python
#새로운 학습 데이터 파일 저장


df.to_csv(feature_file)
```
